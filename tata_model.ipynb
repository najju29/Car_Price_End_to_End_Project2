```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pickle
```

```python
df=pd.read_csv('Tata.csv')
y=df['selling_price']
```

```python
df.head()
```

```python
df.drop('Car_Company',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
model_names=list(df['Model_Name'].unique())
```

```python
len(model_names)
```

```python
avg_prices=[]
for i in model_names:
    avg_prices.append(df[df['Model_Name']==i]['selling_price'].mean())
```

```python
models_df=pd.DataFrame({'Model_name':model_names,'Avg_selling_price':avg_prices})
```

```python
models_df
```

```python
models_dict={}
```

```python
for i in range(0,144):
    models_dict.update([(model_names[i],avg_prices[i])])
```

```python
models_dict
```

```python
Newf_df=df.copy()
```

```python
Newf_df.head()
```

```python
Newf_df.info()
```

```python
Newf_df['Model_Name']=Newf_df['Model_Name'].map(models_dict)
```

```python
Newf_df.head()
```

```python
def fuel_input(a):
    if a=='Petrol':
        return 0
    elif a=='CNG':
        return 1
    elif a=='Diesel':
        return 2
    elif a=='LPG':
        return 3
    
Newf_df['fuel']=Newf_df['fuel'].apply(fuel_input)
```

```python
def seller_input(a):
    if a=='Individual':
        return 0
    elif a=='Dealer':
        return 1
    elif a=='Trustmark Dealer':
        return 2
    
Newf_df['seller_type']=Newf_df['seller_type'].apply(seller_input)
```

```python
def trans_input(a):
    if a=='Manual':
        return 0
    elif a=='Automatic':
        return 1
    
Newf_df['transmission']=Newf_df['transmission'].apply(trans_input)
```

```python
def owner_input(a):
    if a=='First Owner':
        return 0
    elif a=='Second Owner':
        return 1
    elif a=='Third Owner':
        return 2
    elif a=='Fourth & Above Owner':
        return 3
    
Newf_df['owner']=Newf_df['owner'].apply(owner_input)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
tata_model=RandomForestRegressor(n_estimators=310,max_depth=7,min_samples_leaf=2,random_state=3)
```

```python
tata_model.fit(Newf_df('selling_price',axis=1),y)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
rand_params={'n_estimators':[270,290,310,320,330,350],'max_depth':[7,8,9,10,11],'min_samples_leaf':[2,3,4,5]}#,'random_state':[1,3,5,10]}
```

```python
rand_grid=GridSearchCV(estimator=RandomForestRegressor(),param_grid=rand_params,cv=5,verbose=2)
```

```python
rand_grid.fit(Newf_df.drop('selling_price',axis=1),y)
```

```python
rand_grid.best_score_
```

```python
rand_grid.best_params_
```
